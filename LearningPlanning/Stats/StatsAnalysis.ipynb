```python
def plot_correlation(category, file):
    fig, axs = plt.subplots(1, 1, figsize=(8, 8))

    colors = np.array([
        [29, 81, 82],
        [78, 193, 196],
        [217, 69, 74],
        
        [226, 185, 0],
        [110, 0, 255],
        [255, 0, 144],
        [255, 67, 0]
    ], dtype=float) / 255.0

    
    df = None
    try:
        df = pd.read_csv(category + "/" + file, low_memory=False)

        for col in df.columns:
            df[col] = pd.to_numeric(df[col], errors ='coerce').fillna(-1).astype('float')
        df.fillna(-1)
    except:
        print("Error loading:", category + "/" + file)

    legend = file[6:-4].replace("_", " ")

    # Plotting counter-examples acquisition
#         y = np.array([[df[col][i] for col in df.columns if col.startswith("CounterExamples_")]
#                       for i in range(len(df)) if i % step == 0], dtype=float)
#         y[y > 10000] = 0
#         y = y[forward:]
#         ymean = y.mean(axis=1)
#         x = np.arange(y.shape[0]) * step + step * forward

#         color = colors[i % len(colors), :]
#         axs[i0].plot(x, ymean, lw=2, label=legend, color=color)
#         if drawregions:
#             ymin = np.percentile(y, 10, axis=1)
#             ymax = np.percentile(y, 90, axis=1)
#             axs[i0].plot(x, ymin, lw=1, color=color, linestyle="dashed")
#             axs[i0].plot(x, ymax, lw=1, color=color, linestyle="dashed")
#             axs[i0].fill_between(x, ymin, ymax, color=color, alpha=0.1)


    p_revs = np.array([[df[col][i] for col in df.columns if col.startswith("PRev_")]
                      for i in range(len(df))], dtype=float).flatten()
    posits = np.array([[df[col][i] for col in df.columns if col.startswith("Pos_")]
                      for i in range(len(df))], dtype=float).flatten()

    pos = np.zeros(len(p_revs))
    neg = np.zeros(len(p_revs))
    cur_pos = 0
    cur_neg = 0
    for i in range(len(p_revs)):
        if (not np.isnan(p_revs[i])) and p_revs[i] != -1 and posits[i] != -1:
            if posits[i] == 0:
                neg[cur_neg] = p_revs[i]
                cur_neg += 1
            else:
                pos[cur_pos] = p_revs[i]
                cur_pos += 1
    pos = pos[:cur_pos]
    neg = neg[:cur_neg]
    
    print(pos.shape, neg.shape)
        
    sns.histplot(neg, bins=25, stat="probability", element="step", kde=True,
                 ax=axs, label="Negative revision events", color=(217/255, 69/255, 74/255))
    sns.histplot(pos, bins=25, stat="probability", element="step", kde=True,
                 ax=axs, label="Positive revision events", color=(60/255, 150/255, 160/255))
    axs.legend(loc="upper left")
    axs.set_xlabel("Revision probability")
    axs.set_ylabel("Frequency of estimations")
    plt.show()

    return fig
```

```python
fig = plot_correlation("colorblocksworld", "stats_Revision_Probability.csv")
fig.savefig("svgs/posneg_dist.svg")
```
